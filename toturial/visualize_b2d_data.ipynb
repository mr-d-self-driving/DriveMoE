## Data Visualization Notebook
This notebook demonstrates how to use the VisualizeAgent to process scenario data and generate trajectory visualization videos.

### Step 1: Import and Initialization
Initialize the agent with your data directory.

```python
from src.utils.visualize import VisualizeAgent

DATA_ROOT = "exp/b2d_action"
# Initialize the agent
agent = VisualizeAgent(b2d_exp_dir=DATA_ROOT)
```

### Step 2: Load Scenario Data
Specify the scenario_id to retrieve all relevant .pkl files from subdirectories.

```python
# Target scenario ID
scenario_id = "Accident_Town05_Route218_Weather10"

# Fetch file paths and scenario metadata
scenario_data = agent.get_scenario_pkl_files(scenario_id)

# Print total frames found
print(f"Scenario: {scenario_id}")
print(f"Total frames to process: {agent.get_length(scenario_data)}")
```

### Step 3: Generate Visualization Video
Project the trajectory coordinates and compile the frames into an .mp4 video.

```python
# Generate the video output in the 'visualize/' folder
agent.generate_video(scenario_data)
```
